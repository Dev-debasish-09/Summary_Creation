```python
import re
def clean_text(text_lines):
  cleaned_lines = []
  for line in text_lines:
    cleaned_line = re.sub(r'[>*]', '', line)
    cleaned_line = re.sub(r'\s+', ' ', cleaned_line).strip()
    cleaned_lines.append(cleaned_line)
  return cleaned_lines
```

```python
def generate_structure_summary(text):

    subjective_part = text.split(".")[0]
    objective_part = text.split(".")[1]
    assessment_part = text.split(".")[2]
    plan_part = text.split(".")[3]
    subjective_summary = summarizer(subjective_part, max_length=50, min_length=5, do_sample=False)[0]['summary_text']
    objective_summary = summarizer(objective_part, max_length=50, min_length=2, do_sample=False)[0]['summary_text']
    assessment_summary = summarizer(assessment_part, max_length=50, min_length=5, do_sample=False)[0]['summary_text']
    plan_summary = summarizer(plan_part, max_length=50, min_length=5, do_sample=False)[0]['summary_text']
    soap_summary = f"""
    SOAP Notes:
    Subjective: {subjective_summary}
    Objective: {objective_summary}
    Assessment: {assessment_summary}
    Plan: {plan_summary}
    """

    return soap_summary


s_summary = generate_structure_summary(new_text)
print(s_summary)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import accuracy_score, f1_score
```

```python

from torch.utils.data import DataLoader, TensorDataset

train_data = TensorDataset(input_ids, attention_masks, labels)
train_dataloader = DataLoader(train_data, batch_size=8, shuffle=True)

for batch in train_dataloader:
    input_ids, attn_masks, labels = batch
    print(input_ids.shape)
```

```python
from transformers import AdamW
from torch.optim.lr_scheduler import StepLR
from torch.nn import CrossEntropyLoss

optimizer = AdamW(model.parameters(), lr=1e-5)
scheduler = StepLR(optimizer, step_size=1, gamma=0.9)
loss_fn = CrossEntropyLoss()
```

```python
from tqdm import tqdm
import torch

epochs = 3
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct_preds = 0
    total_preds = 0

    for batch in tqdm(train_dataloader):

        input_ids, attention_masks, labels = batch
        # Forward pass
        outputs = model(input_ids, attention_mask=attention_masks, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Calculate accuracy
        _, predicted = torch.max(logits, dim=1)
        correct_preds += (predicted == labels).sum().item()
        total_preds += labels.size(0)

        running_loss += loss.item()


    avg_loss = running_loss / len(train_dataloader)
    accuracy = correct_preds / total_preds
    print(f"Epoch {epoch+1}, Loss: {avg_loss:.4f}, Accuracy: {accuracy:.4f}")


    scheduler.step()
```

```python
def preprocess_text(text):
    inputs = tokenizer(text, padding='max_length', truncation=True, max_length=512, return_tensors="pt")
    return inputs


inputs = preprocess_text(new_text)


with torch.no_grad():
    outputs = model(**inputs)

logits = outputs.logits
predicted_class = torch.argmax(logits, dim=1).item()

label_map = {0: "Negative", 1: "Neutral", 2: "Positive"}
print("Sentiment:", label_map[predicted_class])
```
